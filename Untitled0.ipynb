```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)

    pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        prob = model.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test, prob)
    else:
        auc = np.nan

    return {
        "Accuracy": accuracy_score(y_test, pred),
        "Precision": precision_score(y_test, pred),
        "Recall": recall_score(y_test, pred),
        "F1": f1_score(y_test, pred),
        "AUC": auc
    }
```

```python
lgbm = LGBMClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=7,
    random_state=42
)

xgbm = XGBClassifier(
    n_estimators=500,
    learning_rate=0.10,
    max_depth=6,
    random_state=42,
    eval_metric="logloss",
    verbosity=0
)

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

ensemble = VotingClassifier(
    estimators=[
        ("lgbm", lgbm),
        ("xgb", xgbm),
        ("rf", rf)
    ],
    voting="soft"
)
```

```python
import pandas as pd
import numpy as np

df_all = pd.concat([X_train_rfe, X_test_rfe], axis=0).reset_index(drop=True)

print("Combined dataset shape:", df_all.shape)
print("\nLabel distribution:")
print(df_all["sample_type"].value_counts())

print("\nTotal ransomware families:")
print(df_all[df_all["sample_type"] == 1]["family_label"].nunique())
```

```python
ransom_family_counts = (
    df_all[df_all["sample_type"] == 1]
    .groupby("family_label")
    .size()
    .sort_values(ascending=False)
)

print(ransom_family_counts)
print("\nTotal ransomware families:", len(ransom_family_counts))
```

```python
eligible_families = ransom_family_counts[ransom_family_counts >= 10].index.tolist()

print("Eligible families with >=10 samples:", len(eligible_families))

np.random.seed(42)
unseen_families = np.random.choice(eligible_families, size=15, replace=False)

print("Selected unseen families:")
print(unseen_families)
```

```python
from sklearn.model_selection import train_test_split

df_ransom = df_all[df_all["sample_type"] == 1].copy()
df_good = df_all[df_all["sample_type"] == 0].copy()

# Ransomware split by family
test_ransom = df_ransom[df_ransom["family_label"].isin(unseen_families)]
train_ransom = df_ransom[~df_ransom["family_label"].isin(unseen_families)]

# Goodware random split
train_good, test_good = train_test_split(
    df_good,
    test_size=0.20,
    random_state=42
)

# Combine ransomware + goodware
train_df = pd.concat([train_ransom, train_good], axis=0).sample(frac=1, random_state=42)
test_df = pd.concat([test_ransom, test_good], axis=0).sample(frac=1, random_state=42)

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

print("\nTrain label distribution:")
print(train_df["sample_type"].value_counts())

print("\nTest label distribution:")
print(test_df["sample_type"].value_counts())

# Family leakage check
train_families = set(train_df[train_df["sample_type"] == 1]["family_label"])
test_families = set(test_df[test_df["sample_type"] == 1]["family_label"])

overlap = train_families.intersection(test_families)

print("\nTrain ransomware families:", len(train_families))
print("Test ransomware families:", len(test_families))
print("Overlapping families:", len(overlap))
print("Overlap:", overlap)
```

```python
meta_cols = ["sample_id", "sample_type", "family_label", "type_label"]

feature_cols = [col for col in df_all.columns if col not in meta_cols]

X_train_zd = train_df[feature_cols]
y_train_zd = train_df["sample_type"]

X_test_zd = test_df[feature_cols]
y_test_zd = test_df["sample_type"]

print("X_train_zd:", X_train_zd.shape)
print("y_train_zd:", y_train_zd.shape)
print("X_test_zd:", X_test_zd.shape)
print("y_test_zd:", y_test_zd.shape)
```

```python
def measure_efficiency(model, model_name, X_train, y_train, X_test, feature_count):
    start = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start

    one_sample = X_test.iloc[:1]

    infer_time = timeit.timeit(
        lambda: model.predict(one_sample),
        number=1000
    ) / 1000 * 1000

    file_name = model_name.replace(" ", "_").replace("-", "_") + ".pkl"
    joblib.dump(model, file_name)

    size_mb = os.path.getsize(file_name) / (1024 * 1024)

    return {
        "Model": model_name,
        "No. of Features": feature_count,
        "Train Time (s)": train_time,
        "Inference Time (ms/sample)": infer_time,
        "Model Size (MB)": size_mb
    }
```

```python
def round_table(df):
    df_round = df.copy()
    for col in df_round.columns:
        if df_round[col].dtype in ["float64", "float32"]:
            df_round[col] = df_round[col].round(4)
    return df_round

table2_final = round_table(table2)
table3_final = round_table(table3)
table4_final = round_table(table4)

display(table2_final)
display(table3_final)
display(table4_final)

table2_final.to_csv("FINAL_Table_II.csv", index=False)
table3_final.to_csv("FINAL_Table_III.csv", index=False)
table4_final.to_csv("FINAL_Table_IV.csv", index=False)
```

```python
# ============================================================
# STEP 1: Load and Verify MLRan Dataset
# ============================================================

import os
import zipfile
import glob
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Check current files
# ------------------------------------------------------------
print("Current directory files:")
print(os.listdir())

# ------------------------------------------------------------
# 2. If mlran-main.zip exists, extract it
# ------------------------------------------------------------
zip_files = glob.glob("*.zip")

print("\nZIP files found:")
print(zip_files)

for z in zip_files:
    if "mlran" in z.lower():
        print(f"\nExtracting: {z}")
        with zipfile.ZipFile(z, 'r') as zip_ref:
            zip_ref.extractall("mlran_extracted")
        print("Extraction complete!")

# ------------------------------------------------------------
# 3. List all CSV / Parquet / Pickle files inside extracted folder
# ------------------------------------------------------------
all_files = []

for ext in ["*.csv", "*.parquet", "*.pkl", "*.pickle"]:
    all_files.extend(glob.glob(f"**/{ext}", recursive=True))

print("\nDataset-related files found:")
for f in all_files:
    print(f)

# ------------------------------------------------------------
# 4. Try to load possible train/test files
# ------------------------------------------------------------
csv_files = [f for f in all_files if f.endswith(".csv")]
parquet_files = [f for f in all_files if f.endswith(".parquet")]

print("\nCSV files:", csv_files)
print("Parquet files:", parquet_files)

# ------------------------------------------------------------
# 5. Load files automatically if names contain train/test
# ------------------------------------------------------------
loaded_data = {}

for f in csv_files:
    name = os.path.basename(f).lower()
    try:
        df_temp = pd.read_csv(f)
        loaded_data[f] = df_temp
        print(f"\nLoaded CSV: {f}")
        print("Shape:", df_temp.shape)
        print("Columns sample:", list(df_temp.columns[:10]))
    except Exception as e:
        print(f"Could not load {f}: {e}")

for f in parquet_files:
    name = os.path.basename(f).lower()
    try:
        df_temp = pd.read_parquet(f)
        loaded_data[f] = df_temp
        print(f"\nLoaded Parquet: {f}")
        print("Shape:", df_temp.shape)
        print("Columns sample:", list(df_temp.columns[:10]))
    except Exception as e:
        print(f"Could not load {f}: {e}")

# ------------------------------------------------------------
# 6. Show all loaded dataset names
# ------------------------------------------------------------
print("\nSuccessfully loaded datasets:")
for i, (name, data) in enumerate(loaded_data.items()):
    print(f"{i}: {name} --> {data.shape}")

# ------------------------------------------------------------
# 7. Inspect possible label/family columns
# ------------------------------------------------------------
for name, data in loaded_data.items():
    print("\n" + "="*80)
    print("File:", name)
    print("Shape:", data.shape)

    print("\nColumns:")
    print(list(data.columns))

    possible_label_cols = [
        col for col in data.columns
        if col.lower() in ["label", "malware", "class", "target", "y", "is_malware", "category"]
    ]

    possible_family_cols = [
        col for col in data.columns
        if "family" in col.lower() or "fam" in col.lower()
    ]

    print("\nPossible label columns:", possible_label_cols)
    print("Possible family columns:", possible_family_cols)

    for col in possible_label_cols:
        print(f"\nValue counts for label column '{col}':")
        print(data[col].value_counts())

    for col in possible_family_cols:
        print(f"\nNumber of unique families in '{col}':", data[col].nunique())
        print("Top 10 families:")
        print(data[col].value_counts().head(10))
```

```python
# ============================================================
# STEP 2: Prepare Clean MLRan Dataset for Experiments
# ============================================================

import os
import pandas as pd
import numpy as np

DATA_DIR = "mlran_extracted/mlran-main/6_experiments/FS_MLRan_Datasets"

# ------------------------------------------------------------
# 1. Load required files again
# ------------------------------------------------------------
X_train_raw = pd.read_csv(os.path.join(DATA_DIR, "MLRan_X_train_RFE.csv"))
X_test_raw  = pd.read_csv(os.path.join(DATA_DIR, "MLRan_X_test_RFE.csv"))

metadata_df = pd.read_csv(os.path.join(DATA_DIR, "mlran_dataset_metadata.csv"))
train_ids_df = pd.read_csv(os.path.join(DATA_DIR, "type_train_sampleids.csv"))
test_ids_df  = pd.read_csv(os.path.join(DATA_DIR, "type_test_sampleids.csv"))

print("Raw shapes:")
print("X_train_raw:", X_train_raw.shape)
print("X_test_raw :", X_test_raw.shape)
print("metadata  :", metadata_df.shape)
print("train_ids :", train_ids_df.shape)
print("test_ids  :", test_ids_df.shape)

# ------------------------------------------------------------
# 2. Show columns briefly
# ------------------------------------------------------------
print("\nX_train_raw columns:")
print(list(X_train_raw.columns[:20]))
print("... total columns:", len(X_train_raw.columns))

print("\nMetadata columns:")
print(list(metadata_df.columns))

print("\nTrain IDs columns:")
print(list(train_ids_df.columns))

print("\nTest IDs columns:")
print(list(test_ids_df.columns))

# ------------------------------------------------------------
# 3. Add split column and combine provided train/test RFE data
# ------------------------------------------------------------
X_train_raw = X_train_raw.copy()
X_test_raw = X_test_raw.copy()

X_train_raw["original_split"] = "train"
X_test_raw["original_split"] = "test"

df_all = pd.concat([X_train_raw, X_test_raw], axis=0, ignore_index=True)

print("\nCombined RFE data shape:", df_all.shape)

# ------------------------------------------------------------
# 4. Identify non-feature columns inside RFE file
# ------------------------------------------------------------
possible_non_feature_keywords = [
    "id", "sample", "sha", "hash", "md5",
    "label", "type", "family", "ransom", "goodware",
    "class", "target", "split"
]

non_feature_cols = []
for col in df_all.columns:
    col_lower = str(col).lower()
    if any(k in col_lower for k in possible_non_feature_keywords):
        non_feature_cols.append(col)

print("\nPossible non-feature columns found in RFE file:")
print(non_feature_cols)

# ------------------------------------------------------------
# 5. Detect label and family columns
# ------------------------------------------------------------
label_col = None
family_col = None

for col in df_all.columns:
    c = str(col).lower()
    if c in ["sample_type", "malware", "label", "target", "class"]:
        label_col = col
    if "family" in c:
        family_col = col

print("\nDetected label column in RFE file:", label_col)
print("Detected family column in RFE file:", family_col)

# ------------------------------------------------------------
# 6. If label/family are not inside RFE file, try metadata merge
# ------------------------------------------------------------
# We first check whether df_all already has sample_type and ransomware_family.
# If not, we need to merge using sample IDs.
if label_col is None or family_col is None:
    print("\nLabel/family not fully found inside RFE file.")
    print("Trying to merge with metadata...")

    # Find possible ID column in df_all and metadata
    def find_id_col(df):
        for key in ["sample_id", "id", "sha256", "sha1", "md5", "hash"]:
            for col in df.columns:
                if str(col).lower() == key:
                    return col
        for col in df.columns:
            if "id" in str(col).lower() or "sha" in str(col).lower() or "hash" in str(col).lower():
                return col
        return None

    df_id_col = find_id_col(df_all)
    meta_id_col = find_id_col(metadata_df)

    print("Detected df_all ID column:", df_id_col)
    print("Detected metadata ID column:", meta_id_col)

    if df_id_col is not None and meta_id_col is not None:
        df_all = df_all.merge(
            metadata_df,
            left_on=df_id_col,
            right_on=meta_id_col,
            how="left",
            suffixes=("", "_meta")
        )
        print("Merge complete. New df_all shape:", df_all.shape)
    else:
        print("Could not auto-detect ID columns.")
        print("If this happens, paste the columns output to me.")

# ------------------------------------------------------------
# 7. Re-detect label and family columns after possible merge
# ------------------------------------------------------------
label_col = None
family_col = None

for col in df_all.columns:
    c = str(col).lower()
    if c == "sample_type":
        label_col = col
    if c == "ransomware_family":
        family_col = col

print("\nFinal label column:", label_col)
print("Final family column:", family_col)

if label_col is None or family_col is None:
    raise ValueError("Label or family column not found. Send me the printed column names.")

# ------------------------------------------------------------
# 8. Create y and family
# ------------------------------------------------------------
y_all = df_all[label_col].astype(int)
family_all = df_all[family_col].astype(str)

print("\nClass distribution:")
print(y_all.value_counts())

print("\nFamily count:", family_all.nunique())
print("Top 15 families:")
print(family_all.value_counts().head(15))

# ------------------------------------------------------------
# 9. Prepare feature matrix
# ------------------------------------------------------------
# Remove metadata/non-feature columns
remove_cols = []

for col in df_all.columns:
    c = str(col).lower()

    if (
        c in ["sample_type", "ransomware_family", "ransomware_type", "type_label"] or
        "sha" in c or
        "md5" in c or
        "hash" in c or
        "family" in c or
        "source" in c or
        "timestamp" in c or
        "date" in c or
        "extension" in c or
        "detection" in c or
        "split" in c or
        "id" == c or
        "sample_id" == c
    ):
        remove_cols.append(col)

# Keep only numeric columns as features
candidate_features = [col for col in df_all.columns if col not in remove_cols]
X_all = df_all[candidate_features].select_dtypes(include=[np.number]).copy()

print("\nNumber of removed non-feature columns:", len(remove_cols))
print("Feature matrix shape:", X_all.shape)

print("\nFirst 20 feature columns:")
print(list(X_all.columns[:20]))

# ------------------------------------------------------------
# 10. Check missing values
# ------------------------------------------------------------
missing_count = X_all.isna().sum().sum()
print("\nMissing values in X_all:", missing_count)

if missing_count > 0:
    X_all = X_all.fillna(0)
    print("Missing values filled with 0.")

# ------------------------------------------------------------
# 11. Check original provided train/test family overlap
# ------------------------------------------------------------
if "original_split" in df_all.columns:
    train_families_original = set(df_all.loc[df_all["original_split"] == "train", family_col])
    test_families_original = set(df_all.loc[df_all["original_split"] == "test", family_col])

    overlap_original = train_families_original.intersection(test_families_original)

    print("\nOriginal provided split family overlap:")
    print("Train families:", len(train_families_original))
    print("Test families :", len(test_families_original))
    print("Overlapping families:", len(overlap_original))
    print("Overlap sample:", list(overlap_original)[:20])

# ------------------------------------------------------------
# 12. Final ready variables
# ------------------------------------------------------------
print("\n" + "="*80)
print("STEP 2 COMPLETE")
print("="*80)
print("Ready variables:")
print("X_all       :", X_all.shape)
print("y_all       :", y_all.shape)
print("family_all  :", family_all.shape)
print("df_all      :", df_all.shape)
```

```python
# ============================================================
# STEP 3: Strict Unseen-Family Split for MLRan
# ============================================================

from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 1. Settings
# ------------------------------------------------------------
RANDOM_STATE = 42
N_UNSEEN_FAMILIES = 15
GOODWARE_NAME = "goodware"

np.random.seed(RANDOM_STATE)

# ------------------------------------------------------------
# 2. Create helper dataframe for splitting
# ------------------------------------------------------------
split_df = pd.DataFrame({
    "index": X_all.index,
    "label": y_all.values,
    "family": family_all.values
})

print("Total samples:", len(split_df))
print("\nOverall class distribution:")
print(split_df["label"].value_counts())

# ------------------------------------------------------------
# 3. Separate goodware and ransomware
# label 0 = goodware, label 1 = ransomware
# ------------------------------------------------------------
goodware_df = split_df[split_df["label"] == 0].copy()
ransomware_df = split_df[split_df["label"] == 1].copy()

print("\nGoodware samples:", len(goodware_df))
print("Ransomware samples:", len(ransomware_df))

# ------------------------------------------------------------
# 4. List ransomware families only
# ------------------------------------------------------------
ransomware_families = sorted(
    ransomware_df["family"].dropna().unique().tolist()
)

# Remove goodware if accidentally present
ransomware_families = [
    f for f in ransomware_families
    if str(f).lower() != GOODWARE_NAME
]

print("\nTotal ransomware families:", len(ransomware_families))

family_counts = ransomware_df["family"].value_counts()
print("\nTop 20 ransomware families:")
print(family_counts.head(20))

# ------------------------------------------------------------
# 5. Select unseen ransomware families
# ------------------------------------------------------------
unseen_families = np.random.choice(
    ransomware_families,
    size=N_UNSEEN_FAMILIES,
    replace=False
)

unseen_families = sorted(list(unseen_families))

print("\nSelected unseen ransomware families:")
print(unseen_families)

print("\nSelected unseen family sample counts:")
print(family_counts.loc[unseen_families].sort_values(ascending=False))

# ------------------------------------------------------------
# 6. Split ransomware by family
# ------------------------------------------------------------
ransomware_test_df = ransomware_df[
    ransomware_df["family"].isin(unseen_families)
].copy()

ransomware_train_df = ransomware_df[
    ~ransomware_df["family"].isin(unseen_families)
].copy()

# ------------------------------------------------------------
# 7. Split goodware 80/20
# ------------------------------------------------------------
goodware_train_df, goodware_test_df = train_test_split(
    goodware_df,
    test_size=0.2,
    random_state=RANDOM_STATE,
    shuffle=True
)

# ------------------------------------------------------------
# 8. Combine train and test
# ------------------------------------------------------------
train_df = pd.concat(
    [ransomware_train_df, goodware_train_df],
    axis=0
).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

test_df = pd.concat(
    [ransomware_test_df, goodware_test_df],
    axis=0
).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

# ------------------------------------------------------------
# 9. Create final train/test matrices
# ------------------------------------------------------------
train_indices = train_df["index"].values
test_indices = test_df["index"].values

X_train_zd = X_all.loc[train_indices].copy()
X_test_zd  = X_all.loc[test_indices].copy()

y_train_zd = y_all.loc[train_indices].copy()
y_test_zd  = y_all.loc[test_indices].copy()

family_train_zd = family_all.loc[train_indices].copy()
family_test_zd  = family_all.loc[test_indices].copy()

# ------------------------------------------------------------
# 10. Verification
# ------------------------------------------------------------
print("\n" + "="*80)
print("UNSEEN-FAMILY SPLIT SUMMARY")
print("="*80)

print("Train shape:", X_train_zd.shape)
print("Test shape :", X_test_zd.shape)

print("\nTrain class distribution:")
print(y_train_zd.value_counts())

print("\nTest class distribution:")
print(y_test_zd.value_counts())

print("\nTrain class percentage:")
print((y_train_zd.value_counts(normalize=True) * 100).round(2))

print("\nTest class percentage:")
print((y_test_zd.value_counts(normalize=True) * 100).round(2))

# ------------------------------------------------------------
# 11. Check ransomware-family overlap only
# ------------------------------------------------------------
train_ransom_families = set(
    family_train_zd[y_train_zd == 1].astype(str)
)

test_ransom_families = set(
    family_test_zd[y_test_zd == 1].astype(str)
)

family_overlap = train_ransom_families.intersection(test_ransom_families)

print("\n" + "="*80)
print("FAMILY OVERLAP CHECK")
print("="*80)

print("Train ransomware families:", len(train_ransom_families))
print("Test ransomware families :", len(test_ransom_families))
print("Overlapping ransomware families:", len(family_overlap))
print("Overlap:", family_overlap)

# ------------------------------------------------------------
# 12. Goodware overlap note
# ------------------------------------------------------------
print("\nGoodware exists in both train and test by design.")
print("Strict holdout is applied to ransomware families only.")

# ------------------------------------------------------------
# 13. Final variables ready
# ------------------------------------------------------------
print("\n" + "="*80)
print("STEP 3 COMPLETE")
print("="*80)

print("Ready variables:")
print("X_train_zd      :", X_train_zd.shape)
print("X_test_zd       :", X_test_zd.shape)
print("y_train_zd      :", y_train_zd.shape)
print("y_test_zd       :", y_test_zd.shape)
print("family_train_zd :", family_train_zd.shape)
print("family_test_zd  :", family_test_zd.shape)
```

```python
# ============================================================
# STEP 4: Baseline Models on Strict Unseen-Family Split
# ============================================================

!pip -q install lightgbm xgboost scikit-learn

import time
import warnings
import numpy as np
import pandas as pd

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    balanced_accuracy_score,
    classification_report,
    confusion_matrix
)

from sklearn.ensemble import RandomForestClassifier, VotingClassifier
import lightgbm as lgb
import xgboost as xgb

warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# 1. Define models
# ------------------------------------------------------------
models = {
    "LightGBM": lgb.LGBMClassifier(
        n_estimators=500,
        max_depth=7,
        learning_rate=0.05,
        random_state=42,
        n_jobs=-1,
        verbose=-1
    ),

    "XGBoost": xgb.XGBClassifier(
        n_estimators=500,
        max_depth=6,
        learning_rate=0.10,
        random_state=42,
        eval_metric="logloss",
        n_jobs=-1,
        verbosity=0
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    )
}

# Soft-voting ensemble
models["Soft Voting Ensemble"] = VotingClassifier(
    estimators=[
        ("lgbm", lgb.LGBMClassifier(
            n_estimators=500,
            max_depth=7,
            learning_rate=0.05,
            random_state=42,
            n_jobs=-1,
            verbose=-1
        )),
        ("xgb", xgb.XGBClassifier(
            n_estimators=500,
            max_depth=6,
            learning_rate=0.10,
            random_state=42,
            eval_metric="logloss",
            n_jobs=-1,
            verbosity=0
        )),
        ("rf", RandomForestClassifier(
            n_estimators=300,
            random_state=42,
            n_jobs=-1
        ))
    ],
    voting="soft"
)

# ------------------------------------------------------------
# 2. Evaluation function
# ------------------------------------------------------------
def evaluate_model(model, model_name, X_train, y_train, X_test, y_test):
    print("\n" + "="*80)
    print(f"Training: {model_name}")
    print("="*80)

    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time

    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        y_prob = y_pred

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    macro_prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    macro_rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    macro_f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    bal_acc = balanced_accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)
    pr_auc = average_precision_score(y_test, y_prob)

    cm = confusion_matrix(y_test, y_pred, labels=[0, 1])

    print(f"Accuracy          : {acc:.4f}")
    print(f"Precision         : {prec:.4f}")
    print(f"Recall            : {rec:.4f}")
    print(f"F1                : {f1:.4f}")
    print(f"Macro-F1          : {macro_f1:.4f}")
    print(f"Balanced Accuracy : {bal_acc:.4f}")
    print(f"ROC-AUC           : {roc_auc:.4f}")
    print(f"PR-AUC            : {pr_auc:.4f}")
    print(f"Train Time        : {train_time:.2f} seconds")

    print("\nConfusion Matrix [labels: 0=Goodware, 1=Ransomware]:")
    print(cm)

    print("\nClassification Report:")
    print(classification_report(
        y_test,
        y_pred,
        target_names=["Goodware", "Ransomware"],
        digits=4,
        zero_division=0
    ))

    result = {
        "Model": model_name,
        "Accuracy": round(acc, 4),
        "Precision": round(prec, 4),
        "Recall": round(rec, 4),
        "F1": round(f1, 4),
        "Macro Precision": round(macro_prec, 4),
        "Macro Recall": round(macro_rec, 4),
        "Macro-F1": round(macro_f1, 4),
        "Balanced Accuracy": round(bal_acc, 4),
        "ROC-AUC": round(roc_auc, 4),
        "PR-AUC": round(pr_auc, 4),
        "Train Time (s)": round(train_time, 2),
        "TN": int(cm[0, 0]),
        "FP": int(cm[0, 1]),
        "FN": int(cm[1, 0]),
        "TP": int(cm[1, 1])
    }

    return model, result

# ------------------------------------------------------------
# 3. Train and evaluate all models
# ------------------------------------------------------------
trained_models = {}
baseline_results = []

for model_name, model in models.items():
    trained_model, result = evaluate_model(
        model,
        model_name,
        X_train_zd,
        y_train_zd,
        X_test_zd,
        y_test_zd
    )

    trained_models[model_name] = trained_model
    baseline_results.append(result)

# ------------------------------------------------------------
# 4. Create final baseline result table
# ------------------------------------------------------------
baseline_table = pd.DataFrame(baseline_results)

baseline_table = baseline_table.sort_values(
    by=["F1", "ROC-AUC"],
    ascending=False
).reset_index(drop=True)

print("\n" + "="*80)
print("FINAL BASELINE MODEL COMPARISON TABLE")
print("="*80)

display(baseline_table)

# ------------------------------------------------------------
# 5. Save table for paper
# ------------------------------------------------------------
baseline_table.to_csv("mlran_unseen_family_baseline_results.csv", index=False)

print("\nSaved result table as:")
print("mlran_unseen_family_baseline_results.csv")

# ------------------------------------------------------------
# 6. Identify best model
# ------------------------------------------------------------
best_model_name = baseline_table.loc[0, "Model"]
best_model = trained_models[best_model_name]

print("\nBest model based on F1 + ROC-AUC:")
print(best_model_name)
```

```python
# ============================================================
# STEP 7: Repeated Unseen-Family Evaluation with 5 Seeds
# ============================================================

import os
import time
import warnings
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    balanced_accuracy_score,
    confusion_matrix
)

from sklearn.ensemble import RandomForestClassifier, VotingClassifier
import lightgbm as lgb
import xgboost as xgb

warnings.filterwarnings("ignore")

os.makedirs("mlran_results", exist_ok=True)

# ------------------------------------------------------------
# 1. Settings
# ------------------------------------------------------------
seeds = [42, 7, 21, 77, 100]
N_UNSEEN_FAMILIES = 15
GOODWARE_NAME = "goodware"

# ------------------------------------------------------------
# 2. Helper function: create unseen-family split
# ------------------------------------------------------------
def create_unseen_family_split(seed):
    np.random.seed(seed)

    split_df = pd.DataFrame({
        "index": X_all.index,
        "label": y_all.values,
        "family": family_all.astype(str).values
    })

    goodware_df = split_df[split_df["label"] == 0].copy()
    ransomware_df = split_df[split_df["label"] == 1].copy()

    ransomware_families = sorted(
        ransomware_df["family"].dropna().unique().tolist()
    )

    ransomware_families = [
        str(f) for f in ransomware_families
        if str(f).lower() != GOODWARE_NAME
    ]

    unseen_families = np.random.choice(
        ransomware_families,
        size=N_UNSEEN_FAMILIES,
        replace=False
    )

    unseen_families = sorted([str(f) for f in unseen_families])

    ransomware_test_df = ransomware_df[
        ransomware_df["family"].isin(unseen_families)
    ].copy()

    ransomware_train_df = ransomware_df[
        ~ransomware_df["family"].isin(unseen_families)
    ].copy()

    goodware_train_df, goodware_test_df = train_test_split(
        goodware_df,
        test_size=0.2,
        random_state=seed,
        shuffle=True
    )

    train_df = pd.concat(
        [ransomware_train_df, goodware_train_df],
        axis=0
    ).sample(frac=1, random_state=seed).reset_index(drop=True)

    test_df = pd.concat(
        [ransomware_test_df, goodware_test_df],
        axis=0
    ).sample(frac=1, random_state=seed).reset_index(drop=True)

    train_indices = train_df["index"].values
    test_indices = test_df["index"].values

    X_train = X_all.loc[train_indices].copy()
    X_test = X_all.loc[test_indices].copy()

    y_train = y_all.loc[train_indices].copy()
    y_test = y_all.loc[test_indices].copy()

    family_train = family_all.loc[train_indices].astype(str).copy()
    family_test = family_all.loc[test_indices].astype(str).copy()

    train_ransom_families = set(family_train[y_train == 1])
    test_ransom_families = set(family_test[y_test == 1])
    family_overlap = train_ransom_families.intersection(test_ransom_families)

    return {
        "X_train": X_train,
        "X_test": X_test,
        "y_train": y_train,
        "y_test": y_test,
        "family_train": family_train,
        "family_test": family_test,
        "unseen_families": unseen_families,
        "family_overlap": family_overlap,
        "train_shape": X_train.shape,
        "test_shape": X_test.shape,
        "train_class_counts": y_train.value_counts().to_dict(),
        "test_class_counts": y_test.value_counts().to_dict()
    }

# ------------------------------------------------------------
# 3. Helper function: define fresh models for each seed
# ------------------------------------------------------------
def get_models(seed):
    models = {
        "LightGBM": lgb.LGBMClassifier(
            n_estimators=500,
            max_depth=7,
            learning_rate=0.05,
            random_state=seed,
            n_jobs=-1,
            verbose=-1
        ),

        "XGBoost": xgb.XGBClassifier(
            n_estimators=500,
            max_depth=6,
            learning_rate=0.10,
            random_state=seed,
            eval_metric="logloss",
            n_jobs=-1,
            verbosity=0
        ),

        "Random Forest": RandomForestClassifier(
            n_estimators=300,
            random_state=seed,
            n_jobs=-1
        )
    }

    models["Soft Voting Ensemble"] = VotingClassifier(
        estimators=[
            ("lgbm", lgb.LGBMClassifier(
                n_estimators=500,
                max_depth=7,
                learning_rate=0.05,
                random_state=seed,
                n_jobs=-1,
                verbose=-1
            )),
            ("xgb", xgb.XGBClassifier(
                n_estimators=500,
                max_depth=6,
                learning_rate=0.10,
                random_state=seed,
                eval_metric="logloss",
                n_jobs=-1,
                verbosity=0
            )),
            ("rf", RandomForestClassifier(
                n_estimators=300,
                random_state=seed,
                n_jobs=-1
            ))
        ],
        voting="soft"
    )

    return models

# ------------------------------------------------------------
# 4. Helper function: evaluate one model
# ------------------------------------------------------------
def evaluate_one_model(model, model_name, seed, split_info):
    X_train = split_info["X_train"]
    X_test = split_info["X_test"]
    y_train = split_info["y_train"]
    y_test = split_info["y_test"]

    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    cm = confusion_matrix(y_test, y_pred, labels=[0, 1])

    result = {
        "Seed": seed,
        "Model": model_name,
        "Train Samples": X_train.shape[0],
        "Test Samples": X_test.shape[0],
        "Train Goodware": int((y_train == 0).sum()),
        "Train Ransomware": int((y_train == 1).sum()),
        "Test Goodware": int((y_test == 0).sum()),
        "Test Ransomware": int((y_test == 1).sum()),
        "Unseen Families": len(split_info["unseen_families"]),
        "Family Overlap": len(split_info["family_overlap"]),
        "Accuracy": round(accuracy_score(y_test, y_pred), 4),
        "Precision": round(precision_score(y_test, y_pred, zero_division=0), 4),
        "Recall": round(recall_score(y_test, y_pred, zero_division=0), 4),
        "F1": round(f1_score(y_test, y_pred, zero_division=0), 4),
        "Macro-F1": round(f1_score(y_test, y_pred, average="macro", zero_division=0), 4),
        "Balanced Accuracy": round(balanced_accuracy_score(y_test, y_pred), 4),
        "ROC-AUC": round(roc_auc_score(y_test, y_prob), 4),
        "PR-AUC": round(average_precision_score(y_test, y_prob), 4),
        "Train Time (s)": round(train_time, 2),
        "TN": int(cm[0, 0]),
        "FP": int(cm[0, 1]),
        "FN": int(cm[1, 0]),
        "TP": int(cm[1, 1]),
        "Unseen Family Names": ", ".join(split_info["unseen_families"])
    }

    return result

# ------------------------------------------------------------
# 5. Run repeated unseen-family experiments
# ------------------------------------------------------------
all_seed_results = []

for seed in seeds:
    print("\n" + "="*100)
    print(f"Running unseen-family split for seed = {seed}")
    print("="*100)

    split_info = create_unseen_family_split(seed)

    print("Train shape:", split_info["train_shape"])
    print("Test shape :", split_info["test_shape"])
    print("Train class counts:", split_info["train_class_counts"])
    print("Test class counts :", split_info["test_class_counts"])
    print("Unseen families:", split_info["unseen_families"])
    print("Family overlap:", split_info["family_overlap"])

    if len(split_info["family_overlap"]) != 0:
        raise ValueError(f"Family overlap found for seed {seed}!")

    seed_models = get_models(seed)

    for model_name, model in seed_models.items():
        print(f"\nTraining {model_name} for seed {seed}...")
        result = evaluate_one_model(model, model_name, seed, split_info)
        all_seed_results.append(result)

        print(
            f"{model_name} | "
            f"F1={result['F1']} | "
            f"ROC-AUC={result['ROC-AUC']} | "
            f"FN={result['FN']} | FP={result['FP']}"
        )

# ------------------------------------------------------------
# 6. Create full repeated-results table
# ------------------------------------------------------------
repeated_results_table = pd.DataFrame(all_seed_results)

print("\n" + "="*100)
print("FULL REPEATED UNSEEN-FAMILY RESULTS")
print("="*100)

display(repeated_results_table)

repeated_results_path = "mlran_results/repeated_unseen_family_results_5_seeds.csv"
repeated_results_table.to_csv(repeated_results_path, index=False)

print("Saved full repeated results:", repeated_results_path)

# ------------------------------------------------------------
# 7. Mean ± STD summary by model
# ------------------------------------------------------------
metric_cols = [
    "Accuracy",
    "Precision",
    "Recall",
    "F1",
    "Macro-F1",
    "Balanced Accuracy",
    "ROC-AUC",
    "PR-AUC",
    "Train Time (s)",
    "FP",
    "FN"
]

mean_table = repeated_results_table.groupby("Model")[metric_cols].mean().round(4)
std_table = repeated_results_table.groupby("Model")[metric_cols].std().round(4)

summary_rows = []

for model_name in mean_table.index:
    row = {"Model": model_name}

    for metric in metric_cols:
        mean_val = mean_table.loc[model_name, metric]
        std_val = std_table.loc[model_name, metric]

        row[metric + " Mean"] = mean_val
        row[metric + " Std"] = std_val

    summary_rows.append(row)

summary_table = pd.DataFrame(summary_rows)

# Sort by F1 mean and ROC-AUC mean
summary_table = summary_table.sort_values(
    by=["F1 Mean", "ROC-AUC Mean"],
    ascending=False
).reset_index(drop=True)

print("\n" + "="*100)
print("MEAN ± STD SUMMARY ACROSS 5 UNSEEN-FAMILY SPLITS")
print("="*100)

display(summary_table)

summary_path = "mlran_results/repeated_unseen_family_summary_mean_std.csv"
summary_table.to_csv(summary_path, index=False)

print("Saved mean/std summary:", summary_path)

# ------------------------------------------------------------
# 8. Paper-friendly compact summary
# ------------------------------------------------------------
paper_summary = summary_table[
    [
        "Model",
        "Accuracy Mean", "Accuracy Std",
        "F1 Mean", "F1 Std",
        "Macro-F1 Mean", "Macro-F1 Std",
        "Balanced Accuracy Mean", "Balanced Accuracy Std",
        "ROC-AUC Mean", "ROC-AUC Std",
        "PR-AUC Mean", "PR-AUC Std",
        "FN Mean", "FN Std"
    ]
].copy()

print("\n" + "="*100)
print("PAPER-FRIENDLY SUMMARY TABLE")
print("="*100)

display(paper_summary)

paper_summary_path = "mlran_results/paper_friendly_repeated_summary.csv"
paper_summary.to_csv(paper_summary_path, index=False)

print("Saved paper-friendly summary:", paper_summary_path)

# ------------------------------------------------------------
# 9. Step complete
# ------------------------------------------------------------
print("\n" + "="*100)
print("STEP 7 COMPLETE")
print("="*100)

print("Generated files:")
print(repeated_results_path)
print(summary_path)
print(paper_summary_path)
```
